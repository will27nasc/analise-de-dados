# Coleta de dados

Coletando dados da tabela PDA_Dados_Cursos_Graduacao_Brasil.csv

```python
import pandas as pd

dcgbr_df = pd.read_csv('https://dadosabertos.mec.gov.br/images/conteudo/Ind-ensino-superior/2022//PDA_Dados_Cursos_Graduacao_Brasil.csv', sep=',', usecols=['NOME_IES', 'CATEGORIA_ADMINISTRATIVA', 'ORGANIZACAO_ACADEMICA', 'NOME_CURSO', 'GRAU', 'MODALIDADE', 'SITUACAO_CURSO', 'MUNICIPIO', 'UF', 'REGIAO'])

dcgbr_df
```

Criando uma variavel com valores da base de dados não contendo campo NOME_IES repetido e com situação ativa.

```python
instituicoes_em_atividade = dcgbr_df[(dcgbr_df.SITUACAO_CURSO != 'Em extinção') & (dcgbr_df.SITUACAO_CURSO != 'Extinto')].drop_duplicates(subset=['NOME_IES'])

instituicoes_em_atividade
```

Fazendo a contagem da CATEGORIA_ADMINISTRATIVA para obter o valor de tipo de administração.

```python
adm_instituicoes = instituicoes_em_atividade['CATEGORIA_ADMINISTRATIVA'].value_counts()

adm_name = instituicoes_em_atividade['CATEGORIA_ADMINISTRATIVA'].drop_duplicates()
```

Criando uma nova tabela com os valores dos tipos de administração das instituições de ensino superior.

```python
tabela_adm = {}
nova_tabela_adm = pd.DataFrame(tabela_adm)

for x in range(len(adm_name)):
    adm = adm_name.iloc[x]
    qto = adm_instituicoes.iloc[x]

    nova_linha = pd.DataFrame({ 'Administrador': [adm], 'Quantidade': [qto] }, index=[x])
    nova_tabela_adm = pd.concat([nova_tabela_adm, nova_linha])

display(nova_tabela_adm)
```

Montando o gráfico dividindo percentual de universidades por administração.

```python
import numpy as np 
import matplotlib.pyplot as plt   
  
adm = nova_tabela_adm['Administrador']
qto = nova_tabela_adm['Quantidade']
  
explode = (0.1, 0.0, 0.1, 0.2, 0.3, 0.4) 
colors = ( 'teal', 'orangeRed', 'brown', 
          'grey', 'lightblue', 'beige') 
wp = { 'linewidth' : 1, 'edgecolor' : 'black' } 
def func(pct, allvalues): 
    absolute = int(pct / 100.*np.sum(allvalues)) 
    return '{:.2f}%'.format(pct, absolute) 
fig, ax = plt.subplots(figsize =(10, 7)) 
wedges, texts, autotexts = ax.pie(qto,  
                                  autopct = lambda pct: func(pct, qto), 
                                  explode = explode,  
                                  labels = adm, 
                                  shadow = True, 
                                  colors = colors, 
                                  startangle = 50, 
                                  wedgeprops = wp, 
                                  textprops = dict(color ="black")) 
ax.legend(wedges, adm, 
          title ='Tipos de administração', 
          loc ='center left', 
          bbox_to_anchor =(0.9, 0.1, 1.0, 0.6)) 
  
plt.setp(autotexts, size = 8, weight ="bold") 
ax.set_title('Percentual iniciativa publica X iniciativa privada - 2022') 
plt.savefig("grafico_01.png")
plt.show() 
```

Montando o gráfico dividindo quantidade de instituições de ensino superior por administração.

```python
plt.title("Quantidade iniciativa publica X iniciativa privada - 2022")
plt.barh(adm, qto, color='lightgreen', height=0.6)
plt.savefig("grafico_02.png", dpi = 80, bbox_inches='tight')
plt.show()
```

Fazendo a contagem da REGIAO para obter o valor de tabela regioes.

```python
regioes = instituicoes_em_atividade['REGIAO'].value_counts()

regioes_name = instituicoes_em_atividade['REGIAO'].drop_duplicates()

regioes
```

Criando uma nova tabela com os valores das regioes.

```python
tabela_regioes = {}
nova_tabela_regioes = pd.DataFrame(tabela_regioes)

for x in range(len(regioes_name)):
    regioes_n = regioes_name.iloc[x]
    qto_r = regioes.iloc[x]

    nova_linha_r = pd.DataFrame({ 'Região': [regioes_n], 'Quantidade': [qto_r] }, index=[x])
    nova_tabela_regioes = pd.concat([nova_tabela_regioes, nova_linha_r])

    
display(nova_tabela_regioes)
```

Excluindo a linha 5, pois não vai ser útil no gráfico.

```python
nova_tabela_regioes.drop(5, axis=0, inplace=True)

nova_tabela_regioes
```

Montando o gráfico dividindo percentual de instituições de ensino superior por região.

```python
rga = nova_tabela_regioes['Região']
qto_rga = nova_tabela_regioes['Quantidade']

fig, ax = plt.subplots(figsize =(10, 7)) 

plt.title("Percentual de instituções de ensino superior por região - 2022")
plt.pie(qto_rga, labels=rga, autopct="%1.2f%%")

plt.legend(bbox_to_anchor=(1.35, 0.7))
plt.savefig("grafico_03.png")
plt.show()
```

Montando o gráfico dividindo quantidade de instituções de ensino superior por região.

```python
plt.title('Quantidade de instituções de ensino superior por região - 2022')
plt.bar(rga, qto_rga, color='orange', width=0.5)
plt.savefig("grafico_04.png")
plt.show()
```

Fazendo a contagem da modalidade para obter o valor da tabela modalidades.

```python
modalidades = instituicoes_em_atividade['MODALIDADE'].value_counts()

modalidades_name = instituicoes_em_atividade['MODALIDADE'].drop_duplicates()
```

Criando uma nova tabela com os valores das modalidades.

```python
tabela_modalidades = {}
nova_tabela_modalidades = pd.DataFrame(tabela_modalidades)

for x in range(len(modalidades_name)):
    moda_n = modalidades_name.iloc[x]
    qto_moda = modalidades.iloc[x]

    nova_linha_m = pd.DataFrame({ 'Modalidade': [moda_n], 'Quantidade por unidades': [qto_moda] }, index=[x])
    nova_tabela_modalidades = pd.concat([nova_tabela_modalidades, nova_linha_m])

    
display(nova_tabela_modalidades)
```

Criando a tabela de porcentagem entre as modalidades

```python
mda = nova_tabela_modalidades['Modalidade']
qto_mda = nova_tabela_modalidades['Quantidade por unidades']

colors = ['lightskyblue', 'gold']

explode = (0.1, 0.0,)  

plt.pie(qto_mda, explode=explode, labels=mda, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

plt.legend(mda, loc="upper right")

plt.axis('equal')
plt.savefig("grafico_06.png")
plt.show()
```

```python
plt.title('Quantidade de modalidades presencial x Ead - 2022')
plt.bar(mda, qto_mda, color='blue', width=0.5)
plt.savefig("grafico_07.png")
plt.show()
```

Fazendo a contagem da Universidades por região.

```python
est_ins = instituicoes_em_atividade['UF'].value_counts()

est_ins
```

Criando tabela de quantidade de universidades por estado.

```python
tabela_qto_est_ins = {}
nova_tabela_qto_est_ins = pd.DataFrame(tabela_qto_est_ins)

for x in range(len(est_ins)):
    qto_ins_est = est_ins.iloc[x]

    nova_linha_u = pd.DataFrame({ 'N° Universidades': [qto_ins_est] }, index=[x])
    nova_tabela_qto_est_ins = pd.concat([nova_tabela_qto_est_ins, nova_linha_u])
    
nova_tabela_qto_est_ins.insert(loc=0, column='Estados', value=['SP', 'MG', 'PR', 'BA', 'RJ', 'GO', 'PE', 'RS', 'CE', 'SC', 'PA', 'MT', 'ES', 'DF', 'MA', 'PI', 'PB', 'MS', 'AL', 'AM', 'RN', 'RO', 'TO', 'SE', 'AP', 'AC', 'RR', 'ZZ'])

display(nova_tabela_qto_est_ins)
```

Criando um gráfico da quantiadde de universidades por estado.

```python
insti_estados = nova_tabela_qto_est_ins['Estados']
qto_insti_estados = nova_tabela_qto_est_ins['N° Universidades']

plt.figure(figsize=(11,5))
plt.title('Quantidade de instituções de ensino superior por estado - 2022')
plt.bar(insti_estados, qto_insti_estados, color='red')
plt.savefig("grafico_05.png")
plt.show()
```

### Criação do arquivo PDF

```python
from fpdf import FPDF

class PDF(FPDF):

    def doc_title(self, label):
        self.set_font("helvetica", "B",16)
        self.cell(0, 10, label, 0, 1, "L")
    
    def doc_text(self, text):
        self.set_font("helvetica", "",12)
        self.multi_cell(0, 7, text)
        self.ln()
    
    def doc_image(self, img, x, y, w, h):
        self.image(img, x, y, w, h)

pdf = PDF()
pdf.add_page()

pdf.doc_title("Dados quantitativos das instituições de ensino superior brasileiras")

pdf.doc_image("university.png", 8, 20, (640/3.3), (480/6))

pdf.ln(80)

pdf.doc_text("Uma formação superior é um passo muito importante na vida de uma pessoa. Ter informações sobre as instituições de ensino superior brasileiras é muito importante para tomada de decisões, por isso, separei alguns dados curiosos nesse artigo. O objetivo não é avaliar instituições de ensino, mas sim, apresentar dados quantitativos sobre instituições publico e privado, modelos de ensino, quantidade de instituições de ensino superior por região e etc.")

pdf.doc_title("Quantidade de instituições de ensino superior publicas x privadas")
pdf.doc_text("A quantidade de instituições de ensino superior privadas é muito superior a de instituições de ensino superior públicas. Abaixo seguem os gráficos com o percentual e a quantidade:")

pdf.doc_image("grafico_01.png", 18, 182, (640/4), (480/4))

pdf.add_page()

pdf.doc_image("grafico_02.png", 10, 10, (640/4), (480/4))

pdf.ln(130)

pdf.doc_title("Quantidade de instituições de ensino superior por região")
pdf.doc_text("Nesses gráficos, temos o percentual e quantidade de instituições de ensino superior por regiões. Nessa conta temos polos presenciais e ead.")
pdf.doc_image("grafico_03.png", 10, 175, (640/4), (480/4))

pdf.add_page()

pdf.doc_image("grafico_04.png", 20, 5, (640/4), (480/4))

pdf.ln(120)

pdf.doc_title("Quantidade de instituições de ensino superior por estado")
pdf.doc_text("Já vimos o percentual de instituições de ensino superior por região, agora vamos ver a quantidade de instituições de ensino superior por estado. Nessa contagem tem polos presencias e ead. Obs o elemento ZZ significa polos fora do país.")
pdf.doc_image("grafico_05.png", 20, 165, (640/4), (480/4))

pdf.add_page()

pdf.doc_title("Percentual de polos Presencias X Eads")
pdf.doc_text("Não é nenhuma novidade o crescimento da modalidade Ead no nosso país. Com a pandemia, essa modalidade ganhou ainda mais destaque. O gráfico a seguir mostra essa grande diferença de unidades em relação aos polos presencias.")
pdf.doc_image("grafico_06.png", 20, 40, (640/4), (480/4))
pdf.doc_image("grafico_07.png", 20, 160, (640/4), (480/4))

pdf.add_page()

pdf.doc_title("Referências")
pdf.doc_text("https://dadosabertos.mec.gov.br/images/conteudo/Ind-ensino-superior/2022//PDA_Dados_Cursos_Graduacao_Brasil.csv")

pdf.output("grafico.pdf")
```
